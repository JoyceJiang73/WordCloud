### *Important step: reshape a 3-D graph to a 2-D one by using rgb2gray (convert it to a black and white graph)

```python
#mask = mask.reshape((wine_mask.shape[0],-1), order='F')
from skimage.color import rgb2gray
grey_mask=rgb2gray(mask)
grey_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
#transformed_mask = np.ndarray((mask.shape[0],mask.shape[1]), np.int32)
transformed_mask = np.ndarray((grey_mask.shape[0],grey_mask.shape[1]), np.int32)

for i in range(len(grey_mask)):
    transformed_mask[i] = list(map(transform_format, grey_mask[i]))
    
```
